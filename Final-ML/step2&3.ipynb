#### 2.2 Monthly Average Temperature Pattern

```python
df = df.sort_values(by='datetime')
df['month'] = df['datetime'].dt.month
```

```python
#Temperature by month 
monthly_avg = df.groupby('month')['temp'].mean()

plt.figure(figsize=(10,5), facecolor='white')
plt.plot(monthly_avg.index, monthly_avg.values, marker='o', color='orange')
plt.title('Average Monthly Temperature in HCM City (2015-2025)', fontsize = 14, fontweight = 'bold')
plt.xlabel('Month', fontsize= 12)
plt.xticks(range(1,13))
plt.ylabel('Average Temperature (°C)', fontsize= 12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Observations**

* The chart illustrates the average monthly temperature in Ho Chi Minh City, which fluctuates between approximately `27.5°C (December) and 30.1°C (April)`.
* After peaking, the temperature drops rapidly in May and June, and then remains at a lower, more stable level `(around 28.3°C - 28.8°C)` throughout the rainy season (May to November).
* The lowest temperatures occur towards the end of the year, specifically in January and December `(around 27.5°C)`.
* The difference between the highest and lowest average monthly temperature is quite small, `approximately 2.6°C`, indicating that Ho Chi Minh City experiences an all-year-round warm climate.

**Analysis**

* Relationship between Temperature and Seasons:
    * Dry Season (December - April): The temperature shows a strong and continuous increase, which aligns with the description of "lack of rainfall and strong solar radiation" causing the hottest conditions. April reaches the temperature peak, marking the hottest phase of the dry season.
    * Rainy Season (May - November): The sharp drop in temperature in May and the subsequent maintenance at a lower level `(28°C - 28.5°C)` reflect the impact of heavy rainfall and high humidity. Frequent rain helps to cool the air temperature compared to the peak dry season, although the high humidity can still result in a "humid and muggy" atmosphere.

* Climate Characteristics: The small annual temperature range and the high average temperature throughout the year `(always above 27°C)` are typical features of a tropical monsoon climate near the equator.

**Conclusion**

The average monthly temperature data confirms the characteristics of a tropical monsoon climate in Ho Chi Minh City with two distinct seasons. The highest temperatures occur at the end of the dry season (April) due to intense solar radiation and lack of rain. Temperatures decrease and stabilize during the rainy season (May - November) due to the cooling effect of abundant rainfall, although the temperature remains high and warm year-round.

#### 2.4 Comparison of Average Temperature between Dry and Rainy Seasons

```python
def get_season(month):
    if 5<= month <= 11:
        return 'Rainy'
    else:
        return 'Dry'
df['season'] = df['month'].apply(get_season)
```

```python
# Tạo DataFrame trung bình để vẽ
season_avg = (
    df.groupby('season')['temp']
    .mean()
    .reindex(['Dry', 'Rainy'])
    .reset_index()   
)
season_avg
```

**Observations**

* The bar chart compares the average temperature between the two defined seasons (Dry Season: December – April; Rainy Season: May – November) in Ho Chi Minh City during the 2015–2025 period.
* The average temperature of the Dry Season is clearly higher than that of the Rainy Season.
* The difference in average temperature between the two seasons is relatively small `(only between 0.5°C and 1°C)`, suggesting that warm conditions persist year-round.

**Analysis**

* Dry Season is Warmer: The higher average temperature during the Dry Season is typical of a tropical monsoon climate, where the dry months (especially March and April) have low cloud cover, minimal rainfall, and intense solar radiation, leading to heat accumulation.
* Rainy Season is Cooler: The slightly lower average temperature during the Rainy Season highlights the role of heavy rainfall and frequent cloud cover. The evaporation of rainwater helps cool the atmosphere, lowering the average temperature, although high humidity levels may still contribute to a muggy sensation.
* Low Thermal Amplitude: The fact that the average temperature difference between the two seasons is not drastic `(only about 1°C)` reinforces the characteristics of an equatorial climate, where seasonal temperature variation is not severe.

**Conclusion**

The seasonal data analysis confirms that the Dry Season is the warmer period in Ho Chi Minh City, which aligns with the climatic description. However, the temperature difference between the Dry and Rainy Seasons is minor, confirming that Ho Chi Minh City maintains a stable, warm climate throughout the year, with effective temperature regulation provided by rainfall during the Rainy Season.


### 3. Relationship between temperature and different features

#### 3.1 Correlation matrix

```python
# Relationship between features
corr = df.corr(numeric_only=True)
plt.figure(figsize= (12,10))
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt=".2f", square=True, cbar_kws={"shrink": .8})
plt.title("Feature Correlation Matrix", fontsize=16, fontweight="bold")
plt.show()
```

**Observation:**

The Feature Correlation Matrix reveals strong and logical relationships among weather variables in Ho Chi Minh City, consistent with a tropical monsoon climate:

- Temperature group (`tempmax`, `temp`, `tempmin`, `feelslike*`) shows extremely high positive correlations (**r > 0.90**), indicating near-perfect internal consistency.
- `solarradiation`, `solarenergy`, `uvindex` are almost perfectly correlated (**r ≈ 1.00**), confirming they measure nearly identical solar exposure phenomena.
- Strong negative correlation between:
  - `humidity` and `temp*` variables (**r ≈ -0.82**): higher temperatures → lower relative humidity.
  - `cloudcover` and solar variables (**r ≈ -0.70 to -0.74**): more clouds → less solar radiation.
- Precipitation-related variables (`precip`, `precipprob`, `precipcover`) are highly inter-correlated (**r > 0.80**), showing strong redundancy.
- `visibility` has moderate positive correlation with solar variables (**r ≈ 0.30–0.34**) and negative with cloudcover/humidity — clear skies improve visibility.
- `month` shows moderate correlations with temperature (**r ≈ 0.38**) and precipitation (**r ≈ 0.39**), reflecting seasonal patterns.
- `snow` and `snowdepth` are completely uncorrelated (NaN) — expected in a tropical region.

**Evaluation:**

→ This matrix is highly informative for feature selection and engineering in ML models:

- Redundant features can be dropped to reduce multicollinearity:
  - Keep only one of: `solarradiation`, `solarenergy`, `uvindex` → e.g., keep `uvindex`.
  - Keep only one temperature proxy → e.g., keep `temp`.
  - Drop `snow`, `snowdepth` entirely.
- Strong predictors identified:
  - `temp` ↔ `humidity`, `solarradiation`, `cloudcover` → core drivers of daily weather.
  - `month` → powerful seasonal encoder (can replace complex cyclical features).
